```python
import pandas as pd
```

# merge des dataframes

```python
def remplacer_par_mode(df):
    mode = df.mode().iloc[0]
    df.fillna(mode, inplace=True)
```

```python
df_customers = pd.read_csv('archive/olist_customers_dataset.csv',low_memory=False)
df_customers=df_customers[['customer_id', 'customer_unique_id','customer_state']]
df_customers
```

```python
df_order =pd.read_csv('archive/olist_orders_dataset.csv',low_memory=False)

mask = df_order['order_status']=='delivered'
df_order=df_order[mask]

df_order
```

```python
list_colum = ['order_purchase_timestamp',
        'order_delivered_carrier_date',
       'order_delivered_customer_date', 'order_estimated_delivery_date']

def transformation_en_date(df,list_colum):
       for col in list_colum:
              df[col] =pd.to_datetime(df[col])
              df[col] = df[col].dt.date
       return df


transformation_en_date(df_order,list_colum)

# df_order['order_status'] = df_order.apply(lambda row: 'delivered' if row['order_status'] =='delivered' else 'no_delivered', axis=1)

df_order['recence'] = df_order['order_purchase_timestamp'].max() - df_order['order_purchase_timestamp']

df_order['delai_livraison'] = df_order['order_delivered_customer_date'] - df_order['order_purchase_timestamp']

df_order['delais_respecter'] = df_order.apply(lambda row: 'oui' if row['order_delivered_customer_date'] < row['order_estimated_delivery_date'] else 'non', axis=1)

remplacer_par_mode(df_order)

df_order['delai_livraison'] = df_order['delai_livraison'].dt.days.astype(int)
df_order['recence'] = df_order['recence'].dt.days.astype(int)

df_order = df_order[['order_id', 'customer_id','recence','delai_livraison', 'delais_respecter']]

df_order
```

```python
df_order.info()
```

```python
df_merged1 = pd.merge(df_order, df_customers, how='left',on='customer_id')
df_merged1
```

```python
df_payments =pd.read_csv('archive/olist_order_payments_dataset.csv',low_memory=False)
df_payments=df_payments[['order_id','payment_installments','payment_value',]]
df_payments = df_payments.groupby('order_id').agg({'payment_value': 'sum','payment_installments': 'sum'})
df_payments = df_payments.reset_index()
df_payments
```

```python
df_merged2 = pd.merge(df_merged1, df_payments, how='left',on='order_id')
df_merged2
```

```python
df_items =pd.read_csv('archive/olist_order_items_dataset.csv',low_memory=False)
df_items=df_items[['order_id','product_id','seller_id','price','freight_value']]
df_items
```

```python
df_items =pd.read_csv('archive/olist_order_items_dataset.csv',low_memory=False)

df_items
```

```python
df_merged3 = pd.merge(df_merged2, df_items, how='left',on='order_id')
df_merged3
```

```python
df_products =pd.read_csv('archive/olist_products_dataset.csv',low_memory=False)
df_products['produit_volume_cm_cube'] = df_products['product_length_cm']*df_products['product_width_cm']*df_products['product_height_cm']
df_products= df_products[['product_id', 'product_category_name','product_weight_g','produit_volume_cm_cube']]

df_products
```

```python
df_merged4 = pd.merge(df_merged3, df_products, how='left',on='product_id')
df_merged4
```

```python
df_traslation =pd.read_csv('archive/product_category_name_translation.csv',low_memory=False)

df_traslation
```

```python
df_merged5 = pd.merge(df_merged4, df_traslation, how='left',on='product_category_name')
df_merged5
```

```python
df_seller =pd.read_csv('archive/olist_sellers_dataset.csv',low_memory=False)
df_seller.columns
df_seller =df_seller[['seller_id','seller_state']]
df_seller
```

```python
df_merged6 = pd.merge(df_merged5, df_seller, how='left',on='seller_id')
df_merged6
```

```python
df_reviews =pd.read_csv('archive/olist_order_reviews_dataset.csv',low_memory=False)
df_reviews.columns
df_reviews =df_reviews[['order_id', 'review_score']]
df_reviews
```

```python
df_merged7 = pd.merge(df_merged6, df_reviews, how='left',on='order_id')
df_merged7
```
